1.3 Напишите функцию, которая по `username` ищет человека и выводит информацию о нем. Если пользователь с заданным `username` отсутствует, возбудите исключение `ValueError`

````python
def find_user(inf):
    try:
        for i in range(len(jofile)):
            if inf in jofile[i]['username']:
                print(jofile[i])
        else:
            raise ValueError
    except ValueError:
        print('not found')
find_user('uhebert')
find_user('jygyi')
find_user('llloookkk')
````

1.4 Посчитайте, сколько мужчин и женщин присутсвует в этом наборе данных.

````python
male =0
female=0
for i in range(len(jofile)):
    if jofile[i]['sex'] == 'F':
        female+=1
    else:
        male+=1
print('M:',male)
print('F:',female)
````
